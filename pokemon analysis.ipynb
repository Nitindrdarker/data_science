# Which pokemon type is the strongest

```python
df1.isnull().sum()
```

so many null values in type 2 

fror this observation we will remove type 2 

```python
df2 = df1.drop(['Type 1', 'Type 2'], axis='columns')
df2
```

```python
df2['Total'] = df2['HP'] + df2['Defense'] + df2['Attack'] + df2['Sp. Atk'] + df2['Sp. Def'] + df2['Speed']
df2
```

```python
df2.sort_values(['Total'], ascending=False)
```

```python
legendary = df1[df2['Legendary'] == True].sort_values('Total' ,ascending=False)
legendary
```

```python
legendary['Type 1'].head(3).unique()
```
